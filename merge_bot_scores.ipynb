```python
import pandas as pd
```

```python
politifact_df = pd.read_csv('politifact.csv')
```

```python
tweets_politifact_df = pd.read_csv('crawling/politifact_han.csv')
```

```python
politifact_df.shape
```

```python
tweets_politifact_df.shape
```

```python
politifact_df = politifact_df.set_index('id')
tweets_politifact_df = tweets_politifact_df.set_index('id')
```

```python
df = tweets_politifact_df.join(politifact_df, on='id', how='left')
```

```python
df
```

```python
df = df[df['text'].notna()]
df.drop_duplicates(subset=['text', "title"], keep=False)
```

```python
df.loc['target'] = df['target'].astype(int)
df
```

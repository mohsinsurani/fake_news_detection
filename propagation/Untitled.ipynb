```python
g1 = poli_gr_arr[0]
g2 = poli_gr_arr[1]
g3 = poli_gr_arr[2]

# we don't need to sort the values since the histogram will handle it for us
deg_g1 = [v for k, v in nx.degree(g1)]
deg_g2 = [v for k, v in nx.degree(g2)]
deg_g3 = [v for k, v in nx.degree(g3)]

# there are smarter ways to choose bin locations, but since
# degrees must be discrete, we can be lazy...
max_degree = max(deg_g1 + deg_g2 + deg_g3)

# plot different styles to see both
fig = plt.figure()
ax = fig.add_subplot(111)

ax.hist(deg_g1, bins=range(0, max_degree), density=True, histtype='bar', rwidth=0.8)
ax.hist(deg_g2, bins=range(0, max_degree), density=True, histtype='step', lw=3) 
ax.hist(deg_g3, bins=range(0, max_degree), density=True, histtype='step', lw=3) 

# setup the axes to be log/log scaled
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('degree')
ax.set_ylabel('relative density')
ax.legend()

plt.show()
```
